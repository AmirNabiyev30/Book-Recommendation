This notebook is to create a json file that will store our mapping to use for our model so on inference time we can pass an ISBN
and get it's positional mapping to use for prediction in the model

We will create the mapping the exact same way we created it for the model which is with pandas .cat.code which is slightly unconvential but I took an AI suggestion to create the mapping

```python
import json
import pandas as pd
import sqlite3

conn = sqlite3.connect('./backend/main.db')

query ="SELECT isbn FROM ratings"

df = pd.read_sql_query(query,conn)
```

```python
print(df['isbn'][:20])
```

```python
#create the mapping dictionary

df["book_id"] = df["isbn"].astype("category").cat.codes
print(df['book_id'][:20])
```

```python
#save this file as a json in the backednd folder

mapping_dict = dict(zip(df['isbn'], df['book_id'])) #zip combines iterables and dict creates a dictionary from them
with open('./backend/isbn_to_bookid_mapping.json','w') as f: #creating the json file with the intention of writing
    json.dump(mapping_dict,f) # dumping the dictionary ino the json file
```
